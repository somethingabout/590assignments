## Obtaining the names of the restaurants in Manhattan    
To obtain all the information available, I must scrape more than one page. I notice that the format of url for different    pages depends on the number of the restaurants that has already displayed in previous pages.   
As an example:     
For the url 'https://www.yelp.com/search?cflt=restaurants&find_loc=New%20York,%20NY'   
the url for page 1 ends with:   

"***&start=0***"          

the url for page 24 eneds with:    

"***&start=230***"       

Using the pattern, I can create a for loop to scrape all the available pages    


***Reference used: ***     
1. Scraping the restaurant names on Yelp: https://www.youtube.com/watch?v=r3-v81c2Oew        
2. How to scrape more than one page using Python: https://www.geeksforgeeks.org/how-to-scrape-multiple-pages-of-a-website-using-python/        
3. Using df.append function to add rows into the existing dataframe: https://www.stackvidhya.com/add-row-to-dataframe/       
4. Scraping Yelp using Python: https://levelup.gitconnected.com/scraping-yelp-data-with-python-and-beautiful-soup-39f9088bf633

Reference: Renaming the columns    
https://stackoverflow.com/questions/11346283/renaming-column-names-in-pandas

```python
# Creating an empty list and name it "yelp"
yelp_m = []
# each page displays 10 results. Then the url for the next page must ends with "&start=" plus a multiple of 10. 
# therefore, I create a range of number starting from 0 to 230, with a difference of 10. (modified from Reference 2)
for number in range(0,281,10):
    # this step obtains the correct url for each loop. For the first loop, the url I am getting ends with "&start=0"
    response_m = requests.get(url_m + '&start=' + str(number))
 
    # storing the response in beautiful soup format, and name it soup. 
    soup_m = BeautifulSoup(response_m.text, 'html.parser')
    # create an index number a, and let it start as 0
    a = 0   
    for item in soup_m.select('[class*=container]'): # The codes below are modified from Reference 4. 
        try:
            #print(item)
            if item.find('h4'): # if the item was found under "h4"
                # find the name of the restaurant and store it as "name"
                name = item.find('h4').get_text()
                # under the class a get "reviewCount", obtain the number of reviews 
                reviews = soup_m.select('[class*=reviewCount]')[a].get_text()
                # under the class a get aria-label, obtain the rating
                ratings = soup_m.select('[aria-label*=rating]')[a]['aria-label']
                # find the price range (displayed in terms of $ sign, $ is the cheapest, $$$$$ is the most expensive)
                price = soup_m.select('[class*=priceRange]')[a].get_text()
                # detailed information of the restaurant
                price_category = soup_m.select('[class*=priceCategory]')[a].get_text()
                # add the items found in the list "yelp". 
                yelp_m.append([name, reviews, ratings, price, price_category]) # modified from Reference 3. 
                # print everything found 
                print(name)
                print(reviews)        
                print(ratings)    
                print(price)
                print(price_category)
                print('------------------')
                # tell python to move on to the next restaurant
                a = a+1
        except:
            print(None)
```

```python
len(yelp_m)
```

# Part 2: Cleaning the Dataset        

### Cleaning the Names of the Restaurants

Turning a list into a dataframe and rename the columns.      
The code below is modified from: https://www.kite.com/python/answers/how-to-build-a-pandas-dataframe-with-a-for-loop-in-python         
Merging dataframes: https://www.geeksforgeeks.org/merge-two-dataframes-with-same-column-names/

```python
# Turning the lists "yelp_m", "yelp_b", "yelp_q", "yelp_bx", and "yelp_s" - into a dataframe and rename all the columns accordingly.
# for Manhattan
yelp_m = pd.DataFrame(yelp_m, columns=["Names", "Number of Reviews", "Rating", "Price Range", "Detail"])
# create a new column "Borough" and fill each row with "Manhattan"
yelp_m["Borough"] = "Manhattan"
# for Brooklyn
yelp_b = pd.DataFrame(yelp_b, columns=["Names", "Number of Reviews", "Rating", "Price Range", "Detail"])
# create a new column "Borough" and fill each row with "Brooklyn"
yelp_b["Borough"] = "Brooklyn"
# for Queens
yelp_q = pd.DataFrame(yelp_q, columns=["Names", "Number of Reviews", "Rating", "Price Range", "Detail"])
# create a new column "Borough" and fill each row with "Queens"
yelp_q["Borough"] = "Queens"
# for Bronx
yelp_bx = pd.DataFrame(yelp_bx, columns=["Names", "Number of Reviews", "Rating", "Price Range", "Detail"])
# create a new column "Borough" and fill each row with "Bronx"
yelp_bx["Borough"] = "Bronx"
# for Staten Island
yelp_s = pd.DataFrame(yelp_s, columns=["Names", "Number of Reviews", "Rating", "Price Range", "Detail"])
# create a new column "Borough" and fill each row with "Staten Island"
yelp_s["Borough"] = "Staten Island"

# combine the above four dataframes into one dataframe
yelp_df = pd.concat([yelp_m, yelp_b, yelp_q, yelp_bx, yelp_s], axis=0)
yelp_df
```

The code below are modified from this reference link: https://datascienceparichay.com/article/pandas-split-column-by-delimiter/

```python
# The purpose is to remove the numbers before the names of the restaurants. 
# Using the split function, I separated the numbers and the names into two columns, and store them in a new dataframe called "yelp_df1"
yelp_df1 = yelp_df['Names'].str.split('.', expand=True)
yelp_df1
# Replace the column "Names" in yelp_df, with the new column in yelp_df1 without the numbers. 
# There is a space at the beginning of the restaurant names. Therefore I slice the names starting from the 2nd item, using the str function.
yelp_df["Names"] = yelp_df1[1].str[1:]
```

```python
# Display yelp_df with only the restaurant names.
yelp_df
```

### Cleaning the Variable: Number of Reviews
As shown below, the format of the variable "Number of Reviews" is in object form. To be able to run regressions, I need to transform them into float type.       
Converting columns into numeric form: https://stackoverflow.com/questions/58211251/attributeerror-series-object-has-no-attribute-to-numeric

```python
print(yelp_df["Number of Reviews"].dtypes)
```

```python
# converting the variable into integer form
yelp_df['Number of Reviews'] = yelp_df['Number of Reviews'].astype(int)
print(yelp_df["Number of Reviews"].dtypes)
```

### Cleaning the Variable: Rating
To be able to run a regression, the variable "Rating" must be transformed into float. Currently, the ratings are displayed in "the number of stars" + "star rating" form. I will slice the string and only keep the numbers portion of the column.       
Converting columns into numeric form: https://stackoverflow.com/questions/58211251/attributeerror-series-object-has-no-attribute-to-numeric         

```python
# removing "star rating" from the column "Rating", by slicing the values up to the 12th item counting from the right. 
yelp_df['Rating'] = yelp_df['Rating'].str[:-12]
# the column "Rating" is still in string form. I use the to_numeric function to convert the column into float. 
yelp_df['Rating'] = pd.to_numeric(yelp_df['Rating'], errors='coerce')
yelp_df
```

### Cleaning the Variable: Price Range
To be able to run a regression, I need to display the variable "Price Range" in numeric forms. To do this, I will count the number of "dollar sign" displayed, so that the price range will display as from 1 to 5.
The code below is modified from this link: https://pandas.pydata.org/docs/reference/api/pandas.Series.str.count.html

```python
yelp_df["Price Range"] = yelp_df["Price Range"].str.count('\\$')
yelp_df
```

### Obtaining the locations of the restaurants from the column "Detail"      
An example of the information in the column "Detail"      

ShanghaineseSeafoodNoodles$$Chinatown      

Separated by the dollar sign, the first part of the string is the category of this restaurant, and the second part is location. To obtain location from this column, I can separate this column by the dollar sign. 

```python
# Split the column by the dollar sign, and store the results in a new dataframe "yelp_df2"
yelp_df2 = yelp_df['Detail'].str.split('$', expand=True)
yelp_df2
```

```python
# As shown above, columns 1-4 are the information I need. They are displayed in four columns, because restaurants that 
# are labelled as "$$$$" will take up more columns with missing values. I will merge these four columns into one. 
# Using the astype(str) function, I will convert columns 1-4 into string. 
# The columns with missing value will be converted into a string called "None."
yelp_df["Location"] = yelp_df2[1].astype(str)+ yelp_df2[2].astype(str) + yelp_df2[3].astype(str) + yelp_df2[4].astype(str)
# Then I must remove the "None"s. I use the replace function to replace "None" with "". In other words all the "None"s are removed. 
# Lastly, I store the results in a new column called "Location" in "yelp_df"
yelp_df["Location"] = yelp_df["Location"].str.replace('None', '')
yelp_df.drop("Detail", axis=1, inplace=True)
yelp_df
```

## Item 1: Create 2 graphs between 2 variables. 

### Graph 1: Rating vs. Price Level of the Restaurant
The variable rating ranges from 0 stars to five stars. Price Level ranges from 1 to 4.       

Reference:           
1. Line of best fit: https://www.kite.com/python/answers/how-to-plot-a-line-of-best-fit-in-python      
2. Changing the color of histogram: https://stackoverflow.com/questions/42172440/python-matplotlib-histogram-color
3. List of color names: https://matplotlib.org/stable/gallery/color/named_colors.html
4. Savefig function: https://stackoverflow.com/questions/30765455/why-is-my-plt-savefig-is-not-working

```python
# the np.ployfit function sets up the line of best fit. "m" represents the slop of a linear relationship between x and y. 
# "b" represents the intercept. The number "1" in the end represents that this is a linear relationship. 
m, b = np.polyfit(yelp['Price Range'], yelp['Rating'], 1)

# plotting a scatter plot between price level and rating and reset the color of the scatter plot. 
plt.plot(yelp['Price Range'], yelp['Rating'], 'o', color = "indigo")

# plot the line of best fit, with price level as the independent variable, and the predictted value is calculated through 
# m*x + b
plt.plot(yelp['Price Range'], m*yelp['Price Range'] + b)

# setting the axis titles and the title
plt.xlabel('Price Level', fontsize=15) 
plt.ylabel('Rating', fontsize=15) 
plt.title("Rating vs. Price Level of Restaurants", fontsize=15)

# save the png. 
plt.savefig('Rating vs. Price Level of Restaurants.png', dpi=300, bbox_inches='tight')

plt.show()
```

## Modifying Graph 1 using Jitter

The variable "price range" and "rating" are both discrete-valued. Plotting a scatter graph above led to overplotting and is not informative. Using jitter, a build-in function of seaborn's regplot, I can move each point slightly away from its true value (Â±0.2 range of their true values).      
The "scatter_kws" parameter changes the transparency of the points, so I can see the density distribution better.      

The following code is modified from: https://julienbeaulieu.gitbook.io/wiki/sciences/programming/data-analysis/data-visualization/bivariate-exploration-of-data/overplotting-transparency-and-jitter

```python
# identify the slope and intercept
m, b = np.polyfit(yelp['Price Range'], yelp['Rating'], 1)
# importing necessary packages
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import seaborn as sb
# plot using the seaborn.replot function. 
# x_jitter and y_jitter specifies how much I want to plots to be moved away from their true values.
# scatter_kws specifies the transparency of the data points. 
sb.regplot(data = yelp, x = 'Price Range', y = 'Rating', fit_reg = False,
           x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha' : 1/3}, color = "indigo")
# plotting the line of best fit
plt.plot(yelp['Price Range'], m*yelp['Price Range'] + b, color = "blue")

# assigning title and axis labels
plt.xlabel('Price Level', fontsize=15) 
plt.ylabel('Rating', fontsize=15) 
plt.title("Rating vs. Price Level of Restaurants", fontsize=15)

# saving the graph
plt.savefig('(with jitter)Rating vs. Price Level of Restaurants.png', dpi=300, bbox_inches='tight')

# showing the graph
plt.show()
```

## Interpreting the Graph             
The line of best fit is slightly upward sloping. It means the rating of restaurants tend to increase as price increases.       
After applying jitter to teh graph, I now see that most restaurants have a price level of 2 and ratings of 4-4.5. 

### Graph 2: Number of Reviews by Price Level


```python
# group the dataframe yelp by price range, so we can divide the data into four categories, one for each price level. 
# sum the number of reviews for each price level, store the results in a new dataframe called g2. 
g2 = pd.DataFrame(yelp['Number of Reviews'].groupby(yelp['Price Range']).sum().reset_index())
g2

# create a bar plot using g2. I devided the number of reviews by 10,000 to make it easier to read. 
plt.bar(g2["Price Range"], g2['Number of Reviews']/10000,color = "indigo")

# setting the axis titles and the title
plt.xlabel('Price Level', fontsize=15) 
plt.ylabel('Total Number of Reviews\n(in ten thousands)', fontsize=15) 
plt.title("Number of Reviews by Price Level", fontsize=15)

# save the graph as png format. See reference 4. 
plt.savefig('Number of Reviews by Price Level.png', dpi=300, bbox_inches='tight')

# present the graph
plt.show()
```

### Interpretation of the Graph

As shown in the graph above, restaurants within the price level of 2 tend to receive the most reviews from its customers. Both the cheapest and the most expensive restaurants have less reviews than the restaurants with affordability in the middle range. 

## Item 3: Display the graphs side by side in one figure.

Reference: combine the seaborn regplot with matplotlib: https://stackoverflow.com/questions/33049884/how-to-plot-2-seaborn-lmplots-side-by-side

```python
# setting the layout of the two plots and the dimension of the combined figure to be 16 by 8. 
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(16, 8))

# the np.ployfit function sets up the line of best fit. "m" represents the slop of a linear relationship between x and y. 
# "b" represents the intercept. The number "1" in the end represents that this is a linear relationship. 
m, b = np.polyfit(yelp['Price Range'], yelp['Rating'], 1)

# defining the first subplot
ax1 = plt.subplot(1, 2,1)

# plotting a scatter plot between price level and rating and reset the color of the scatter plot. 
sb.regplot(data = yelp, x = 'Price Range', y = 'Rating', fit_reg = False,
           x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha' : 1/3}, color = "indigo", ax=ax1) # see reference 1

# plot the line of best fit, with price level as the independent variable, and the predictted value is calculated through 
# m*x + b
ax1.plot(yelp['Price Range'], m*yelp['Price Range'] + b)

# setting the axis titles and the title
plt.xlabel('Price Level', fontsize=15) 
plt.ylabel('Rating', fontsize=15) 
plt.title("Rating vs. Price Level of Restaurants", fontsize=15)

# defining the second subplot
ax2 = plt.subplot(1, 2,2)

# group the dataframe yelp by price range, so we can divide the data into four categories, one for each price level. 
# sum the number of reviews for each price level, store the results in a new dataframe called g2. 
g2 = pd.DataFrame(yelp['Number of Reviews'].groupby(yelp['Price Range']).sum().reset_index())
g2

# create a bar plot using g2. I devided the number of reviews by 10,000 to make it easier to read. 
ax2.bar(g2["Price Range"], g2['Number of Reviews']/10000,color = "indigo")

# setting the axis titles and the title
plt.xlabel('Price Level', fontsize=15) 
plt.ylabel('Total Number of Reviews\n(in ten thousands)', fontsize=15) 
plt.title("Number of Reviews by Price Level", fontsize=15)


# save the graph as png format. See reference 4. 
plt.savefig('Combined.png', dpi=300, bbox_inches='tight')

# present the graph
plt.show()
```

## Item 4: Additional Graph: Distribution of the Number of Reviews 
Reference:              
1. Setting the number of bins: https://thinkingneuron.com/how-to-visualize-data-distribution-of-a-continuous-variable-in-python/              
2. Changing the color of histogram: https://stackoverflow.com/questions/42172440/python-matplotlib-histogram-color

```python
# plot the histogram of the number of reviews, setting the number of bins to 15 and reset the color. See reference 1-3. 
plt.hist(yelp['Number of Reviews'], bins = 15, color = "indigo", ec="indigo")

# set the title and the axis titles
plt.xlabel('Number of Reviews', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title("Distribution of the Number of Reviews", fontsize=15)

# save the graph as png format. See reference 4. 
plt.savefig('NOR.png', dpi=300, bbox_inches='tight')

# present the graph
plt.show()
```

Reference:        
1. Converting values to natural log: https://www.kite.com/python/answers/how-to-compute-the-natural-logarithm-(ln)-of-a-number-in-python


```python
# The distribution number of reviews is skewed to the right. Therefore I chose to put it on a log scale. 
# Create a new column computing the natural log of the number of reviews. 
yelp["Log of Number of Reviews"] = np.log(yelp["Number of Reviews"])

# Plot the histogram of the log values of the number of reviews. 
plt.hist(yelp["Log of Number of Reviews"], bins = 15, color = "indigo", ec="indigo")

# Set the title and axis titles
plt.xlabel("Natural Log of Number of Reviews", fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title("Distribution of the Number of Reviews", fontsize=15)

# Save as png
plt.savefig('NOR_Log.png', dpi=300, bbox_inches='tight')

# present the graph
plt.show()
```
